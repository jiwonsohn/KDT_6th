##### 결측치 처리 fillna, dropna 메서드
- fillna() - 결측치를 특정 값을 채우기<br>
           - method='ffill' : 결측치 이전 값(행??)으로 대체<br>
           - method='bfill' : 결측치 이후 값(행??)으로 대체<br>
- dropna() - 결측치를 포함하는 행 삭제<br>

           - how='all'          :  특정 행(열) 모든 값이 결측치인 경우에만 삭제<br>

##### 데이터 분위수를 구하는 quntile 메서드

```python
df = pd.read_csv(r'C:\Users\KDP-43\Desktop\KDT_Pandas\data_analysis.Chaneui\datasets\bookings\Bookings.csv')


df['Total_Review'] = df['Total_Review'].map( lambda x: str(x).replace('external','').strip() )
df['Total_Review'] = df['Total_Review'].map( lambda x: str(x).replace('review','').strip() )
df['Total_Review'] = df['Total_Review'].map( lambda x: str(x).replace(',','').strip() )

df['Total_Review'] = df['Total_Review'].astype('float')
```

```python
quantile = [0, 0.2, 0.4, 0.6, 0.8, 1]

for idx in quantile:
    q = df['Total_Review'].quantile( idx, interpolation='lower')
    print(f'quantile( {idx} ) is {q}')
```

```python

for idx in quantile:
   q = df['Total_Review'].quantile( idx, interpolation='nearest')
   print(f'quantile( {idx} ) is {q}')
```

##### 그룹별 연산을 위한 groupby 메서드
- 특정 변수의 그룹별 연산 편리성 증진
- 그룹별 데이터 개수,평균,합계 & 임의 정의 함수 연산 가능


```python
df = sns.load_dataset('titanic')
df.head(10)
```

```python
df.groupby('sex')['survived'].mean()
```

```python
# 그룹으로 묶을 변수가 2개 이상 --> [ var1, var2 ]
df.groupby( ['sex', 'class'])['survived'].mean()
```

```python
# groupby 그룹 연산 데이터에 2개 이상 메서드&함수 연산
#                                                 --> agg( [ 'method1', 'func1' ] )
df.groupby( ['sex','class'])['survived'].agg( ['mean', 'count'])
```

##### 컬럼 별로 연산 함수 다르게 지정
                             --> agg( {'col1' : 'method1', 'col2' : 'func1'})

```python
df.groupby( ['sex', 'class'])[ ['survived', 'age'] ].agg( {'survived':'mean', 'age':'max'})
```

##### groupby + apply() --> 사용자 정의 함수 그룹 연산

```python
import numpy as np

def get_IQR(data):
    _3rd = data.quantile(.75)
    _1st = data.quantile(.25)
    return (np.abs(_3rd - _1st) * 1.5)

df.groupby( ['sex','class'])['age'].apply(get_IQR)
```

##### resample 메서드
- rule인자    : 샘플링 구간 전달 인자
- label 인자  : rule 선언 시간 범위 설정 (left: 앞선, right: 뒷선)

```python
# 데이터셋을 일주일 단위로           --> 7d
# 각 데이터들은 해당 일주일의 평균    --> mean()
# 을 나타내도록 다운 샘플링

df = pd.read_csv(r'C:\Users\KDP-43\Desktop\KDT_Pandas\data_analysis.Chaneui\datasets\APPL_price\APPL_price.csv')
df['Date'] = pd.to_datetime( df['Date'])
df = df.set_index('Date')

df.head(10)
```

```python
df.resample(rule='7d', label='left').mean()
```

```python
df.resample(rule='7d', label='right').mean()
```
