##### fare 하위 50% - '저비용' &  상위 50% - '고비용' 그룹화에 따른 생존 경향성

```python
# fare 그룹 나누기 (fare 중위값 기준)

fare_50 = titanic_cat['fare'].median()

low_fare = titanic_cat['fare'] <= fare_50

import numpy as np
titanic_cat['fare_range'] = np.nan

titanic_cat.loc[low_fare, 'fare_range'] = 'Low_Fare'
titanic_cat.loc[~(low_fare), 'fare_range'] = 'High_Fare'

titanic_cat
```

```python
# 저비용 그룹의 생존 경향

low_fare_0 = sum( titanic_cat[low_fare]['survived'] == 0) / sum(titanic_cat['survived']==0) * 100
low_fare_1 = sum( titanic_cat[low_fare]['survived'] == 1) / sum(titanic_cat['survived']==1) * 100


# 고비용 그룹의 생존 경향

high_fare_0 = sum( titanic_cat[~(low_fare)]['survived'] == 0) / sum(titanic_cat['survived']==0) * 100
high_fare_1 = sum( titanic_cat[~(low_fare)]['survived'] == 1) / sum(titanic_cat['survived']==1) * 100


# print
print(f'fare가 {fare_50:4.2f} 이하인 저비용 그룹의 사망자 비율은 사망자 그룹의 {low_fare_0:4.2f}%')
print(f'fare가 {fare_50:4.2f} 이하인 저비용 그룹의 생존자 비율은 생존자 그룹의 {low_fare_1:4.2f}%')
print()
print(f'fare가 {fare_50:4.2f} 초과인 고비용 그룹의 사망자 비율은 사망자 그룹의 {high_fare_0:4.2f}%')
print(f'fare가 {fare_50:4.2f} 초과인 고비용 그룹의 생존자 비율은 생존자 그룹의 {high_fare_1:4.2f}%')
```

```python
''' 
fare가 높을수록 사망률보다 생존률이 더 증가하는 경향을 확인

저비용 그룹은 사망자 비율이 생존자 비율의 약 2배인 반면,
고비용 그룹은 생존자 비율이 사망자 비율의 약 1.7배

흥미로운 점은 fare가 낮은 그룹의 사망자 & 생존자 비율과
             fare가 높은 그룹의 생존자 & 사망자 비율이 서로 반대라는 점
         
'''
```
