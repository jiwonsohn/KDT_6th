```python
# 모듈 로딩
import pandas as pd
import numpy as np
```

```python
# 인덱스, 칼럼 확인
print(f'인덱스:  {dataDF.index}')
print("="*50)
print(f'컬럼:  {dataDF.columns}')
```

####   1) 신재생 에너지 항목 북한 추가

```python
dataDF.loc[9] = '-'
dataDF.loc[9, "전력량 (억㎾h)"] = np.nan
print(dataDF)
```

```python
dataDF.loc[9, '발전 전력별'] = "신재생"
print(dataDF)
```

####  2) '-' ==> 0 변환

```python
for row in dataDF.index:
    for col in dataDF.columns:
        
        if dataDF.loc[row, col] == "-":
            dataDF.loc[row, col] = 0

        else:
            pass
       
print(dataDF)
```

####   3) 2016년 열 옆에 항목별 합계 칼럼 추가 값은 / 함수나 for 구문으로 sum값

```python
# 합계 칼럼 추가

dataDF.loc[:,"합계"] = 0
print(dataDF)
```

```python
# 각 행에 대한 합계 추가

def df_sum(row, dataframe):
    return sum(dataframe.loc[row, dataframe.columns[2:-1]])

for row in dataDF.index:
    dataDF.loc[row, "합계"] = df_sum(row, dataDF)

print(dataDF)
        
```
