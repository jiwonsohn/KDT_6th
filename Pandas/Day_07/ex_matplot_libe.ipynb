##### [Matplotlib 활용 시각화]
- matplotlib.pyplot 모듈 사용
- 다양한 그래프 관련 함수들 & 클래스들 존재 <br><br>

- plot 기본: 선그래프
- plot(x축데이터, y축데이터, fmf(포맷스트링) )<br>
             --> fmf: 그래프 색상, 두께, 스타일 등등 지정<br>

 예시) plt.plot(  [10,20,30], [5,10,15], 'bs--') <br>

- # 그래프 부가적 기능

# 그래프 제목,  x축, y축
# 범례(legend)

```python
import matplotlib.pyplot as plt
```

```python
# 그래프 그리기
# plot([y축 데이터])

plt.plot([1,2,3,4])
plt.xlabel("DATA")
plt.ylabel("RESULT")
plt.show()
```

```python
# 선 그래프 그리기
# plot(x축데이터, y축데이터)

plt.plot(  [10,20,30], [5,10,15])
```

```python
# 선 그래프 그리기
# plot(x축데이터, y축데이터, fmf(포맷스트링) )

plt.plot(  [10,20,30], [5,10,15], 'ro--')
```

```python
plt.plot(  [10,20,30], [5,10,15], 'bs--')
```

```python
# 선 그래프 그리기
# plot(x축데이터, y축데이터, Line2D 설정으로)

plt.plot( [10,20,30], [5,10,15], color='blue',
         marker='s', linestyle='dashed', markersize=12)
```

```python
# mec = marker edge color
# mew = marker edge width

plt.plot( [10,20,30], [5,10,15], color='blue',
         marker='s', linestyle='dashed', markersize=12,
         mec='magenta', mew=3)

```

```python
# 선 그래프 그리기
# plot(x축데이터, y축데이터, Line2D 설정으로)

plt.plot( [10,20,30], [15,25,35], 'bo-')
plt.plot( [10,20,30], [18,29,25], 'ro--')
plt.show()
```

```python
# 위와 같은 결과 한줄로
plt.plot( [10,20,30], [15,25,35], 'bo-',
         [10,20,30], [18,29,25], 'ro--')

plt.show()
```

```python
# 그래프 부가적 기능

# 그래프 제목,  x축, y축
# 범례(legend) --> 그래프의 label 속성 값을 범례에서 사용

plt.plot([10,20,30],[15,25,20], 'bo--', label="V")
plt.plot([10,20,30],[5,30,28], 'ro--', label="2")

plt.title("tutorial")
plt.xlabel("Data")
plt.ylabel("Result")
plt.legend()
plt.show()
```

```python
# 그래프 부가적 기능

# 그래프 제목,  x축, y축
# 범례(legend) --> 그래프의 label 속성 값을 범례에서 사용
#              --> loc 속성ㅇ로 위치 지정 가능 (기본값: best)

plt.plot([10,20,30],[15,25,20], 'bo--', label="V")
plt.plot([10,20,30],[5,30,28], 'ro--', label="2")

plt.title("tutorial")
plt.xlabel("Data")
plt.ylabel("Result")
plt.legend(loc='center right')
plt.show()
```
