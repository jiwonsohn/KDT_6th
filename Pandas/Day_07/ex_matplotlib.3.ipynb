##### [Matplotlib 활용 데이터 시각화]
- X축 & Y축 다양한 처리

```python
from matplotlib import pyplot as plt
```

```python
# 그래프 그리기 - 기본

y_data = [1,3,5,7,9]
x_data = ['2024-01-01', '2024-02-01', '2024-03-01', '2024-04-01', '2024-05-01']
plt.plot(x_data, y_data, marker='o')
plt.show()
```

```python
# 그래프 그리기 - x축 y축 범위 설정
# only int,float형으로만 범위 설정 가능
# 범위 설정 후, x축&y축 이름 바꾸기!

y_data = [1,3,5,7,9]
x_data = ['2024-01-01', '2024-02-01', '2024-03-01', '2024-04-01', '2024-05-01']
plt.plot(x_data, y_data, marker='o')
plt.xlim(left='2023-12-01', right='2024-07-01')
plt.ylim(bottom=0, top=15)
plt.show()
```

```python
# 그래프 그리기 - x축 y축 눈금(thick) 설정

y_data = [1,3,5,7,9]
x_data = ['2024-01-01', '2024-02-01', '2024-03-01', '2024-04-01', '2024-05-01']
plt.plot(x_data, y_data, marker='o')
plt.xticks(rotation=45)
plt.show()
```

```python
# 그래프 그리기 - 그래프 출력 창, figure 크기 설정

y_data = [1,3,5,7,9]
x_data = ['2024-01-01', '2024-02-01', '2024-03-01', '2024-04-01', '2024-05-01']

plt.figure(figsize=(10, 10))
plt.plot(x_data, y_data, marker='o')
plt.xticks(rotation=45)
plt.show()
```

```python
# 그래프 그리기 - 그래프 출력 창, figure 크기 자동 설정

y_data = [1,3,5,7,9]
x_data = ['2024-01-01', '2024-02-01', '2024-03-01', '2024-04-01', '2024-05-01']

# plt.figure(figsize=(10, 10))
plt.plot(x_data, y_data, marker='o')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 그래프 그리기 - 눈금(ticks) 설정

y_data = [1,3,5,7,9]
x_data = ['2024-01-01', '2024-02-01', '2024-03-01', '2024-04-01', '2024-05-01']

# plt.figure(figsize=(10, 10))
plt.plot(y_data, marker='o')
plt.xticks([11,12,13,14,15])
plt.show()
```

```python
# 그래프 그리기 - 눈금(ticks) 설정

y_data = [1,3,5,7,9]
x_data = ['2024-01-01', '2024-02-01', '2024-03-01', '2024-04-01', '2024-05-01']

# plt.figure(figsize=(10, 10))
plt.plot(y_data, marker='o')
plt.xlim(left=0, right=7)
plt.xticks([0,2,4,6] , ['A', 'B', 'C', 'D'])
plt.show()
```

```python
# 그래프 그리기 - 눈금(ticks) 설정

y_data = [1,3,5,7,9]
x_data = ['2024-01-01', '2024-02-01', '2024-03-01', '2024-04-01', '2024-05-01']

# plt.figure(figsize=(10, 10))
plt.plot(y_data, marker='o')
plt.xlim(left=-1, right=7)
plt.xticks([0,2,4,6] , ['A', 'B', 'C', 'D'], fontweight='bold')
plt.show()
```

```python
# 그래프 그리기 - x축 폰트 설정값 Dict 변수 선언으로

y_data = [1,3,5,7,9]
x_data = ['2024-01-01', '2024-02-01', '2024-03-01', '2024-04-01', '2024-05-01']

# plt.figure(figsize=(10, 10))
plt.plot(y_data, marker='o')
plt.xlim(left=-1, right=7)

# x축 폰트 설정값 Dict 변수 선언
font_config = {'fontweight' :'bold'}
plt.xticks([0,2,4,6] , ['A', 'B', 'C', 'D'], **font_config)
plt.show()
```
