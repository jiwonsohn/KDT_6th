#### [Pie 그래프]
* 컬럼(변수)의 데이터의 분포 비율을 시각화

```python
# 모듈로딩
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 한글패치
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
```

```python
# 데이터 준비
file_path = r'C:\Users\KDP-43\Desktop\KDT_Pandas\Day_09\auto_mpg.csv'
dataDF = pd.read_csv(file_path)
```

```python
# 컬럼별 타입 확인
dataDF.dtypes
```

##### 특정 컬럼의 데이터 비율
- 자동차 생산국가에 대한 비율

```python
# origin 칼럼은 생산지에 대해 숫자로 '범주화'한 값 --> category 타입으로 형 변환
dataDF['origin'].astype('category')
```

```python
# 자동차 생산국가의 전체에 대한 비율
# originSR = round( dataDF['origin'].value_counts() / dataDF['origin'].count(), 2)

originSR = dataDF['origin'].value_counts()

x = originSR.index.tolist()
h = originSR.values.tolist()

x,h
```

##### 자동차 생산국가의 전체에 대한 비율 

##### Bar

```python
# bar
plt.bar( x,h, width=0.6, alpha=0.5, label='자동차 생산국가 비율')

plt.xticks([], [])

```

```python
# bar plot
originbar = plt.bar( x,h, width=0.6, alpha=0.5, color = ['r','b','k'] , edgecolor=['magenta'])

plt.title('국가별 자동차 생산 비율', fontsize=16)
plt.xticks([1,2,3], ['US', 'Japan', 'Europe'], rotation=-45)
plt.xlabel("생산국가", labelpad=10)
plt.ylabel("(%)")

# 바 상단 수치 출력
plt.bar_label(originbar, h, label_type='center')

```

##### pie

```python
plt.pie(h, labels=['US', 'Japan', 'Europe'], autopct='%.1f%%',
        counterclock=False)

plt.show()
```

```python
# 조각 1개 떨어뜨리기
plt.pie(h, labels=['US', 'Japan', 'Europe'], autopct='%.1f%%',
        counterclock=False, explode=[0,0,0.8])

plt.show()
```

```python
# nested pie chart
plt.pie(h, labels=['US', 'Japan', 'Europe'], autopct='%.1f%%',
        counterclock=False, 
        wedgeprops={'linewidth':5, 'width':0.8, 'edgecolor':'w'})

plt.show()
```

```python
# nested pie chart
plt.pie(h, labels=['US', 'Japan', 'Europe'], autopct='%.1f%%',
        counterclock=False, 
        wedgeprops={'linewidth':10, 'width':0.2, 'edgecolor':'r'})

plt.show()
```
