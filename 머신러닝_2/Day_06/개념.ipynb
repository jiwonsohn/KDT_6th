### 데이터 탐색적 분석(EDA)
	-> 모든 것을 수치화 (ex. 범주형: 인코딩, )
	-> 지도/비지도/강화		
	-> 학습 알고리즘 선택

### 선형모델		
	-> 피쳐와 타겟의 관계: 선형 관계 가정
	-> 정규분포 기반
	-> 선형식:	y = ax+b => wx+b => w0 + w1x
			y = w0 + w1x + w2x + ... + wnx

	-> Coefficient or Weight
	: 학습을 통해 다양한 피쳐들의 영향 정도를 정확하게 모델에 
	반영(학습)하기 위해 사용됨

	-> 릿지, 라쏘, 엘라스틱
	: 각 피쳐의 가중치에 규제를 더하기 위한 방법 / 모든 피쳐를 모델에 학습시킴으로써 발생할 수 잇는 과대적합 방지 목적
	: 릿지 => 2 norm
	: 라쏘 => 1 norm

#### 파라미터
* 모델 파라미터
	- fit() 진행 후 설정/정의되는 파라미터
	- 모델명.파라미터명_

* 하이퍼 파라미터
	- 모델 인스턴스 생성 시, 설정/정의 되는 매개변수
	- 개발자가 값을 설정
	- 모델 성능에 영향을 미침
	- 예시)	규제 강도(alpha), K 갯수, max_iter(학습횟수), 규제 강도(C; Cost, 1/alpha)

* 튜닝
	- 모델 성능 향상을 위한 작업
	- 모델의 하이퍼 파라미터의 값을 조절해서 성능 올리기

* 좋은 모델이란?
	- 일반화가 잘 된 모델
	- 모델 복잡도 => 간소한 모델 (피쳐 수가 적은 것)
